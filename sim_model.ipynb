```python
df["cmc"] = df["cmc"].astype("int64")
```

```python
df_dataset = df[df["attraction_lights"].isna()][["id", "name", "mana_cost", "cmc", "type_line", "oracle_text", "colors", "color_identity", "keywords", "set", "rarity", "power", "toughness", "produced_mana", "card_faces", "loyalty"]]
```

```python
df_dataset.head()
```

```python
import json
```

```python
with open("catalog.json") as file:
    catalog = json.load(file)
```

```python
catalog
```

```python
def supertype_col(row):
    vals = []
    for val in catalog["supertypes"]:
        if val in row["type_line"]:
            vals.append(val)
    return vals
```

```python
df_dataset["supertypes"] = df_dataset.apply(supertype_col, axis=1)
df_dataset
```

```python
def cardtypes_col(row):
    vals = []
    for val in catalog["card_types"]:
        if val in row["type_line"]:
            if val == "Plane": continue
            vals.append(val)
    return vals
```

```python
df_dataset["card_types"] = df_dataset.apply(cardtypes_col, axis=1)
df_dataset
```

```python
def subtypes_col(row):
    vals = []
    for val in catalog["subtypes"]:
        if val in row["type_line"]:
            vals.append(val)
    return vals
```

```python
df_dataset["subtypes"] = df_dataset.apply(subtypes_col, axis=1)
df_dataset
```
